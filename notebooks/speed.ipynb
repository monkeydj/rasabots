```python
# referenceed from caromball.com
speeds = np.arange(0.05, 3.6, step=0.5)
print('speeds=', str(speeds[:2]) + '...' + str(speeds[-2:]), 'm/s')
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

t = np.linspace(1, 3, 32)
for s in np.nditer(speeds):
    ax.plot(t, s / t, label=f'v= {s} m/s')

ax.set(xlabel='Time [s]', ylabel='Acceleration [m/s^2]')
ax.legend()

plt.show()
```
